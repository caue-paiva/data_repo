```python
import pandas as pd
import os
import sklearn

df = pd.read_csv(os.path.join('olym_treated_NoNaN_train.csv'))

df.index = df['ID']
df.drop('ID',axis=1,inplace=True)
df.drop('Name',axis=1,inplace=True)
df.head()
```

```python
#functions for parsing
def Set_Sex_binary(df):  #0 is for males and 1 for females 
   Bina_sex = lambda x : int(x == 'F')  
   df['Sex'] =  df['Sex'].apply(Bina_sex)
   return df
```

```python
df2 = pd.read_csv(os.path.join('olymp_train_OG.csv'))

df2 = Set_Sex_binary(df2)
#df2["Sex"].unique()

df["Sex"] = df2["Sex"].copy()
#df.head()
df["Sex"].unique()

```

```python
#categorical data encoder from sklearn
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

categorical_columns =['Team','NOC','City','Sport','Event','Medal']
print(df.shape)
flag = True
for col in categorical_columns:
     df[col] = encoder.fit_transform(df[col]) 
     #print(encoded_df.head())
     #df[col] = encoded_df.columns
     #df = df.drop(col, axis=1)
     #df = pd.concat([df, encoded_df], axis=1)
df.head()
```
