```python
import seaborn as sns
import pandas as pd 
import os 
import matplotlib.pyplot as plt
df = pd.read_csv(os.path.join('olymp_LabelEnco__newest_treated.csv'))

print(df['Sex'].unique())
```

```python


plt.figure(figsize=(15, 15))

sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={"shrink": 0.75})

plt.savefig("heatmap_olymp.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(15, 15))
sns.pairplot(df)
plt.savefig("pairplot_olymp.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
from sklearn.metrics import mutual_info_score

mi_matrix = pd.DataFrame(index=df.columns, columns=df.columns)

for col1 in df.columns:
    for col2 in df.columns:
        mi_matrix.loc[col1, col2] = mutual_info_score(df[col1], df[col2])

mi_matrix = mi_matrix.astype(float)

# Plot the mutual information values using a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(mi_matrix, annot=True, cmap='YlGnBu')
plt.title('Mutual Information between Features')
plt.show()
```
